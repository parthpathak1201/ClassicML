# ClassicML — K-Means
Cluster synthetic blobs using the C++ `KMeans` binding.

```python
import sys
sys.path.insert(0, '../python')
import numpy as np
import matplotlib.pyplot as plt
import classicml_py as cml

rng = np.random.default_rng(7)
centers = np.array([[0, 0], [5, 5], [-4, 4]])
X = np.vstack([c + rng.normal(0, 0.6, (80, 2)) for c in centers])

km = cml.KMeans(3)
km.fit(X)
labels = km.predict(X).astype(int)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10', s=25, edgecolors='k')
plt.title('K-Means clusters (ClassicML C++)')
plt.xlabel('x1'); plt.ylabel('x2')
plt.show()
```
