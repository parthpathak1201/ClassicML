# ClassicML — Linear Regression
Train C++ `LinearRegression` via the Python bridge and plot predictions.

```python
import sys
sys.path.insert(0, '../python')
import numpy as np
import matplotlib.pyplot as plt
import classicml_py as cml

rng = np.random.default_rng(42)
X = rng.uniform(0, 5, (60, 1))
y = 3.0 * X.squeeze() + rng.normal(0, 0.4, 60)

model = cml.LinearRegression()
model.fit(X, y, epochs=200)
pred = model.predict(X)

plt.scatter(X, y, label='data', alpha=0.6)
idx = np.argsort(X.squeeze())
plt.plot(X.squeeze()[idx], pred[idx], 'r-', label='C++ predict')
plt.title('Linear Regression (ClassicML C++)')
plt.legend()
plt.show()
```
