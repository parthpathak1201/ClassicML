# ClassicML — Logistic Regression
Binary classification with moons-like synthetic data.

```python
import sys
sys.path.insert(0, '../python')
import numpy as np
import matplotlib.pyplot as plt
import classicml_py as cml

rng = np.random.default_rng(0)
n = 200
t = rng.uniform(0, np.pi, n)
X = np.column_stack([np.cos(t), np.sin(t)]) + rng.normal(0, 0.15, (n, 2))
y = (t > np.pi / 2).astype(float)

model = cml.LogisticRegression()
model.fit(X, y, epochs=300)
pred = model.predict(X)

acc = np.mean(pred == y)
print(f'Accuracy: {acc:.3f}')

plt.scatter(X[:, 0], X[:, 1], c=pred, cmap='coolwarm', edgecolors='k', s=20)
plt.title('Logistic Regression predictions (ClassicML C++)')
plt.xlabel('x1'); plt.ylabel('x2')
plt.show()
```
